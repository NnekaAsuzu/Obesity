```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import scipy
import PIL
import requests
```

```python
sns.set_style('darkgrid')
fig = plt.subplots(figsize=(10,10))

sns.lineplot(data= obesity,
            x='NObeyesdad',
            y='Gender')
```

```python
#do you eat high calorie food frequently - FAVC

daysplot = sns.relplot(
    data=obesity,
    x="family_history_with_overweight",
    y="NObeyesdad",
    hue="Gender",
    col="FAVC",
    kind="line",
    col_wrap=2
)
```

```python
sns.countplot(data=obesity, x='NObeyesdad')
plt.xticks(rotation=45)
plt.title("Distribution of Obesity Levels")
```

```python
sns.boxplot(data=obesity, x='NObeyesdad', y='Weight')
plt.xticks(rotation=45)
plt.title("Weight by Obesity Level")
```
