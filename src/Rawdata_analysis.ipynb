```python
# Import all the Libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import scipy
import PIL
import requests
```

```python
# set a random seed for reproducibility
np.random.seed(103)
obesity= pd.read_csv('/Users/rohitmalhotra/Desktop/ObesityDataSet_raw_and_data_sinthetic.csv')

#view th etop few lines of the data
print (obesity.head(5))
```

```python
# look at the column names 
col = obesity.columns
print(col)
```

```python
# View the gender vs obesity levels on a lineplot
sns.set_style('darkgrid')
fig = plt.subplots(figsize=(10,10))

sns.lineplot(data= obesity,
            x='NObeyesdad',
            y='Gender')
```

```python
#Check relation between FAVC(do you eat high calorie food frequently),obesity levels, family history with overweight as per gender

daysplot = sns.relplot(
    data=obesity,
    x="family_history_with_overweight",
    y="NObeyesdad",
    hue="Gender",
    col="FAVC",
    kind="line",
    col_wrap=2
)
```

```python
# lineplots werent very clear so will now check the countplots
# to get a better sense of the data distribution , look at the distribution of obesity levels across the data
sns.countplot(data=obesity, x='NObeyesdad')
plt.xticks(rotation=45)
plt.title("Distribution of Obesity Levels")
```

```python
# establishing relation between weight and obesity, with boxplot
sns.boxplot(data=obesity, x='NObeyesdad', y='Weight')
plt.xticks(rotation=45)
plt.title("Weight by Obesity Level")
```

```python
#using an accessible colour palette

sns.boxplot(data=obesity, x='NObeyesdad', y='Weight', palette="colorblind")
plt.xticks(rotation=45)
plt.title("Weight by Obesity Level")
```
